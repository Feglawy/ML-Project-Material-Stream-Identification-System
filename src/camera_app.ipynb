```python
import joblib
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler

model_path = "../models/svm_model.pkl"

data_loaded = joblib.load(model_path)
svm : SVC = data_loaded["model"]
scaler: StandardScaler = data_loaded["scaler"]
```

```python
import cv2 as cv
from feature_extraction import feature_extraction

THRESHOLD = 0.6

class_name = {
     0 :"glass",
     1:"paper",
     2:"cardboard",
     3:"plastic",
     4:"metal",
     5:"trash"
}

cap = cv.VideoCapture(0)  # 0 = default camera

if not cap.isOpened():
    print("Cannot open camera")
    exit()

while True:
    ret, frame = cap.read()
    
    if not ret:
        print("Can't receive frame (stream end?)")
        break

    cv.imshow("Video", frame)
    
    features = feature_extraction(frame)
    features = features.reshape(1, -1) 
    features = scaler.transform(features)
    
    # class_label = svm.predict(features)
    # probabilty = svm.predict_proba(features)

    probs = svm.predict_proba(features)[0]
    class_id = probs.argmax()
    confidence = probs[class_id]


    if confidence < THRESHOLD:
        class_name_str = "unknown"
    else:
        class_name_str = class_name[class_id]



    print(f"{class_name_str} ({confidence:.2f})")
        

    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
